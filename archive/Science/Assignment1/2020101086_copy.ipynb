## Task 0: Simulate a random walk for n timesteps

Make a function that takes n as an input and runs a 1-D random walk for n timesteps, and returns us the
coordinates of the point at all timesteps. Also plot the graph of position x vs t.

```python
# Take n as input
n = int(input("Enter the number of points: "))

# Function to generate random points
def generate_points(n):
    array = np.random.randint(0,2,n+1)
    array = array*2 - 1
    array[0] = 0
    array = np.cumsum(array)
    return array


# Plot graph x vs t
array = generate_points(n)
plt.plot(range(n+1), array)
plt.xlabel('t')
plt.ylabel('x')
plt.show()

```

## Task 1: Simulate a random walk with two people

Two drunks start out together at the origin, each having equal probability of making a step to the left or right
along the X-axis. They make their steps simultaneously. Simulate the experiment appropriate number of times
and answer the following parts:

• Part 1: Find the Probability that they meet again after N timesteps by running the simulation multiple
times and plot the probability for different values of N .
• Part 2: Simulate the random walk of a single drunk person and find the probability that he/she returns
back to the Origin after N timesteps. Plot the probabilities for different values of N .
• Part 3: Find the Mean and Mean Squared Displacement for the drunk after N steps. Perform the
experiment multiple times and find these central tendencies for N time steps.

```python
n = input("Enter the range of N: ")
n = int(n)


points_1 = [ generate_points(n) for i in range(1000) ]
points_2 = [ generate_points(n) for i in range(1000) ]

meet = np.array([ points_1[i] == points_2[i] for i in range(1000) ])
probability = np.sum(meet, axis=0)/1000

# Graph
plt.plot(range(n+1), probability)
plt.xlabel("N")
plt.ylabel("Probability")
plt.show()
```

```python
n = input("Enter the range of N: ")
n = int(n)

points = np.array([ generate_points(n) for i in range(1000) ])
origin = (points==0)
probability = np.sum(origin, axis=0)/1000

# Graph
plt.plot(range(n+1), probability)
plt.xlabel("N")
plt.ylabel("Probability")
plt.show()

```

```python
n = input("Enter the range of N: ")
n = int(n)

points = np.array([ generate_points(n) for i in range(1000) ])
mean = np.mean(points, axis=0)
mean_squared = np.mean(points**2, axis=0)

plt.plot(range(n+1), mean, color = 'red')
plt.plot(range(n+1), mean_squared, color = 'blue')
plt.legend(['Mean', 'Mean Squared'])
plt.xlabel('N')

plt.show()
```

## 2.1 Task

Consider a unit circle and unit square. You have been given N point pebbles. Simulate an experiment to determine the value of π using these inputs. And write a code to find π.

```python
radius = 1

n = input("Enter the range of N: ")
n = int(n)

x = np.random.uniform(0,radius,n)
y = np.random.uniform(0,radius,n)
circle = (x**2 + y**2 <= radius**2)

pi = 4*np.cumsum(circle)/np.arange(1,n+1)

print('The value of pi is: ', pi[-1])

plt.plot(range(1,n+1), pi)

plt.xlabel('Number of points')
plt.ylabel('Value of pi')

plt.show()
```
