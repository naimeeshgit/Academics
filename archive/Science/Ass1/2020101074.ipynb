```python
from numpy import random
import numpy as np
from matplotlib import pyplot as plt
```

### TASK 0

here we simulate random walk using random.random()

```python
n = 100 
list = []
# int(input())
x=0


for i in range(n):
    step = random.random()
    if step < 0.5:
        x = x-1
    else:
        x = x+1
    # list.append(x)
    list.append(x)
print(list)

```

### TASK 1

```python
list1 = []
list2 = []
for n in range(0, 100):
    count =0
    for t in range(1, 1000): 
        xd1 = 0
        xd2 = 0
        for steps in range(1, n+1):

            step1 = random.random()
            step2 = random.random()

            if step1 < 0.5:
                xd1 = xd1-1
            else:
                xd1 = xd1+1

            if step2 < 0.5:
                xd2 = xd2-1
            else:
                xd2 = xd2+1

        if xd1 == xd2:
            count = count + 1

    prob = count/1000
    list1.append(prob)
    list2.append(n)
  

# plot
plt.plot(list2, list1)
plt.show()
```
