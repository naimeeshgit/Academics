```python
from numpy import random
import numpy as np
from matplotlib import pyplot as plt
```

### TASK 0

here we simulate random walk using random.random()

```python
n = 100 

def random_walk(n):
    options = [-1,1]
    first_step = np.zeros(1)
    steps = np.append(first_step, np.random.choice(options,n-1))
    position = np.cumsum(steps)
    return position
```

### TASK 1.1

```python
def thousand_random_walksks(n):
    position =  np.array([ random_walk(n) for i in range(1000) ])
    return position

probability_meet = []
for i in range(1,n+1):
    drunkard_1 = thousand_random_walksks(i)[:,-1]
    drunkard_2 = thousand_random_walksks(i)[:,-1]
    probability_meet.append(((drunkard_1 == drunkard_2).sum())/1000)

# plot
plt.plot(range(n), probability_meet)
plt.show()
```

### TASK 1.2

```python
walks = thousand_random_walksks(n)
occurences = np.equal(walks, 0).sum(axis=0)
probability = occurences/1000

# plot
plt.plot(range(n), probability)
plt.show()
```

### Task 1.3

```python
walks = thousand_random_walksks(n)
mean_disp = walks.mean(axis=0)
mean_square_disp = walks.var(axis=0) + mean_disp**2

print(mean_disp[-1])
print(mean_square_disp[-1])

plt.plot(range(n), mean_disp)
plt.plot(range(n), mean_square_disp)
plt.show()
```
