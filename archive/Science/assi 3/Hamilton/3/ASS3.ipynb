```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Position vs time
    

```python
# Using Hamiltonian equation mx'' = -k(x) to get position vs time
k = (2020101047%3+1) # 3
m= (2020101047 % 5) + 1 # 3
x0= (2020101047 % 7) + 1 # 7
#x0 act as both initial point and amplitude
total_time= 12.566370614359172953850573533118
n=1000

h=total_time/n
T=np.linspace(0, total_time, n+1)#getting n+1 coordinates of t
#For A
Ax = np.zeros((n+1, n+1))
Ax[0, 0] = 1
Ax[n, n] = 1
for i in range(1, n):
    Ax[i, i-1] = m
    Ax[i, i] = -2*m+k*h**2
    Ax[i, i+1] = m    
# Get b
bx = np.zeros(n+1)
bx[0]=x0
bx[n]=x0
X = np.linalg.solve(Ax, bx) #getting n+1 coordinates of x
plt.plot(T,X)
plt.ylabel("position")
plt.xlabel("time")
plt.show()
for i in range(n+1):
    if(X[i]>x0 ):
        X[i]=x0
    if(X[i]<-x0 ):
        X[i]=-x0

        

    
    
```

<h2>Momentum vs Time

```python
#Since total energy is constant, therefore p^2/2m +(1/2)*k*x**2 = (1/2)*k*x0**2 => p =sqrt(mk(x0^2-x^2))
P = np.zeros(n+1)
t=-1
for i in range(n+1):
    P[i]=t*(np.sqrt(m*k*(x0**2 - X[i]**2)))
    if(P[i]==0 and i!=0):
        t=-t
        
plt.plot(T,P)
plt.ylabel("momentum")
plt.xlabel("time")
plt.show()
```

<h2> Minimizing Potential Energy

```python
#Here we are going to minimise potential energy U=(1/2)*k*x**2
delx=0.1 # if using gradient descent the change 'α' (x'=x0-α∇) is constant. While in steepest descent method this alpha is calulated by minimising f(x') 
vector = x0
diff =10
num_iter=100
U=np.zeros(num_iter)
x_coor=np.zeros(num_iter)


def gradient(x):#∂U/∂x=k*x
    return (k*x)
    
for i in range(num_iter):
    U[i]=(1/2)*k*(vector**2)
    x_coor[i]=vector
    diff = -delx * gradient(vector)
    vector += diff
print(vector)
```

<h2> Plot of Position vs Potential Energy

```python
Uideal =1/2*k*(X**2)
plt.plot(X,Uideal,label="Ideal")
plt.scatter(x_coor, U,color='r', label="Calculated")
plt.xlabel("Position")
plt.ylabel("Potential Energy")
plt.legend()
```
