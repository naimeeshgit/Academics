# Class Assignment 6: Integration using Monte Carlo

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## $$ \int_{0}^{1}3x^{2}dx $$

```python
def integral_1D(x0,x1,N,iter):          # Monte-Carlo method for function in one variable
    calc = []
    lim = x1-x0                         # Range for integration
    for i in range(iter):
        x = np.arange(x0,x1,0.01)       # Generating tightly spaced random number between the limit 
        y = max(3*x**2)         # Considering the max of the function
        x_in = x0 + np.random.random(N)*lim     # generating random values for both x and y
        y_in = 0 + np.random.random(N)*y
        d = 3*x_in**2
        
        ind_in = np.where(y_in <= d)            # Checking for which points fall in the integral
        area = y * lim * len(ind_in[0])/N       # Finding out the integral
        calc.append(area)

    return calc

```

### Plot of I v/s points(N)

```python
calc = []
for i in range(1,10000):
    calc.append(integral_1D(0,1,i,1))
calc = np.array(calc)
calc = calc.flatten()
plt.plot(calc)
plt.xlabel("N")
plt.ylabel("$ \int_{0}^{1}3x^{2}dx $")
plt.title("I v/s N")
plt.show()
```

### N=20, 100 trials, Plot of I v/s Trials

```python
calc = integral_1D(0,1,20,100)
plt.plot(calc)
plt.xlabel("Numeber of trials")
plt.ylabel("$ \int_{0}^{1}3x^{2}dx $")
plt.title("I v/s N (N = 20), 100 trials")
plt.show()
```

### Standard Deviation with N=20, 100 trials

```python
print("Standard deviation for N = 20 (100 trials): ",np.std(calc))
```

### N = 1000, 100 trials, Plot I v/s Trials

```python
calc = integral_1D(0,1,1000,100)
plt.plot(calc)
plt.xlabel("N")
plt.ylabel("$ \int_{0}^{1}3x^{2}dx $")
plt.title("I v/s N (N = 1000), 100 trials")
plt.show()
```

### Standard Deviation with N=20, 100 trials

```python
print("Standard deviation for N = 20 (100 trials): ",np.std(calc))
```

### Standard deviation proportional to sqrt(N)

```python
std_dev = []
for i in range(1,1000):
    calc = integral_1D(0,1,i,100)
    std_dev.append(np.std(calc))

N = np.array(list(range(1,1000)))
sq = 1/np.sqrt(N)

plt.plot(std_dev,'b',label='Standard Deviation (N=1000,trials=100)')
plt.plot(sq,'r',label='1/srt(N)')
plt.xlabel('N')
plt.title("Standard deviation and sqrt(N)")
plt.legend()
plt.show()
```

From the plot of standard deviation we obeserve that standard deviation is inversely proportional to sqrt(N)

## $$ \int_{0}^{1}\int_{0}^{1}x^{2}y \space dxdy $$

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def integral_2D(x0,x1,y0,y1,N,iter):            # Monte-Carlo method for function in two variables
    calc = []
    lim_x = x1-x0                               # Range for integration
    lim_y = y1-y0               
    for i in range(iter):
        x = np.arange(x0,x1,0.01)               # Generating tightly spaced random number between the limit
        y = np.arange(y0,y1,0.01)
        z = max(x**2*y)                         # Considering the max of the function
        x_in = x0 + np.random.random(N)*lim_x   # generating random values for both x, y and z
        y_in = y0 + np.random.random(N)*lim_y
        z_in = 0 + np.random.random(N)*z
        d = (x_in**2*y_in)
        
        ind_in = np.where(z_in <= d)                    # Checking for which points fall in the integral
        calc.append(z * lim_x * len(ind_in[0])/N)       # Finding out the integral

    return calc        
```

### Plot of I v/s points(N)

```python
calc = []
for i in range(1,10000):
    calc.append(integral_2D(0,1,0,1,i,1))
calc = np.array(calc)
calc.flatten()
plt.plot(calc)
plt.xlabel("N")
plt.ylabel("$ \int_{0}^{1}\int_{0}^{1}x^{2}y  dxdy $")
plt.title('I v/s N')
plt.show()
```

### N = 20, 100 trials, Plot of I v/s Trials

```python
calc = integral_2D(0,1,0,1,20,100)
plt.plot(calc)
plt.xlabel("Number of Trials")
plt.ylabel("$ \int_{0}^{1}\int_{0}^{1}x^{2}y  dxdy $")
plt.title("I v/s N (N = 20), 100 trials")
plt.show()
```

### Standard Deviation with N=20, 100 trials

```python
print("Standard Decviation for N = 20, (100 trials): ",np.std(calc))
```

### N = 1000, 100 trials, Plot I v/s Trials

```python
calc = integral_2D(0,1,0,1,1000,100)
plt.plot(calc)
plt.xlabel("Number of Trials")
plt.ylabel("$ \int_{0}^{1}\int_{0}^{1}x^{2}y  dxdy $")
plt.title("I v/s N (N = 1000), 100 trials")
plt.show()
```

### Standard Deviation with N=1000, 100 trials

```python
print("Standard Decviation for N = 1000, (100 trials): ",np.std(calc))
```

### Standard deviation proportional to sqrt(N)

```python
std_dev = []
for i in range(1,1000):
    calc = integral_2D(0,1,0,1,i,500)
    std_dev.append(np.std(calc))

N = np.array(list(range(1,1000)))
sq = 1/np.sqrt(N)

plt.plot(std_dev,'b',label='Standard Deviation (N=1000,trials=100)')
plt.plot(sq,'r',label='1/srt(N)')
plt.xlabel('N')
plt.title("Standard deviation and sqrt(N)")
plt.legend()
plt.show()
```

From the plot of standard deviation we obeserve that standard deviation is inversely proportional to sqrt(N)
