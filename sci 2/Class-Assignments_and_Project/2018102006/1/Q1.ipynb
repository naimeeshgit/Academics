# Class Assignment 1: Random Walk

```python
from random import seed
from random import randint
import matplotlib.pyplot as plt
import math
```

```python
def combination(N):             # Finding combinatorics of 2N things taken N at a time
  L1 = [i for i in range(2*N,N,-1)]
  L2 = [j for j in range(N,0,-1)]
  val = 1
  for k in range(len(L1)):
    val *= L1[k]/(4*L2[k])

  return val
```

```python
seed(1)

N = int(input("Enter the total no.of steps: "))     # User input for total number of steps
w1 = 0                  # Initializing the variables for computation 
w2 = 0                  # w1, w2, are the walkers
c1 = 0                  # c1 for checking how many times the two meet
c2 = 0                  # c2 for checking how many times walker 1 ends up at the origin
x = []
y1 = []
a1 = []
y2 = []
w1_mean = [0 for m in range(N)]         # Arrays for storing the values of mean and mean squared displacement
w2_mean = [0 for m in range(N)]
w1_meansq = [0 for m in range(N)]
w2_meansq = [0 for m in range(N)]

for i in range(1,N+1):          # Implementing random walk model. Checking for the meeting point of the two walkers after 
  x.append(i)                   # N steps
  c1 = 0
  c2 = 0
  for j in range(100):          # 100 trials
    w1 = 0
    w2 = 0
    for k in range(i):
      val1 = randint(-1,1)
      while (val1 == 0):
        val1 = randint(-1,1)
      w1 += val1
      val2 = randint(-1,1)
      while (val2 == 0):
        val2 = randint(-1,1)
      w2 += val2

    w1_meansq[i-1] += w1**2         # Doing the necessary computations
    w2_meansq[i-1] += w2**2
    if (w1 == w2):                  # Updating the counts
      c1 += 1
    if (w1 == 0):
      c2 += 1
    
    w1_mean[i-1] += w1
    w2_mean[i-1] += w2
    
  w1_mean[i-1] /= i*100             # Final computation of the arrays 
  w2_mean[i-1] /= i*100  
  w1_meansq[i-1] /= 100
  w2_meansq[i-1] /= 100

  y1.append(c1/100)                 #Array for count after 100 trials
  y2.append(c2/100)
  a1.append(combination(i))
```

```python
plt.plot(x,y1,'r', label='Computational Solution')
plt.plot(x,a1,'b', label='Analytical Solution')
plt.xlabel('N')
plt.ylabel('P(two drunks meet after N steps)')
plt.title('Two 1-D Random Walkers')
plt.legend()
plt.show()    
```

```python
plt.plot(x,y2,'r', label='Computational Solution')
plt.xlabel('N')
plt.ylabel('P(Returning to origin after N steps)')
plt.title('P_Origin')
plt.legend()
plt.show()  
```
