# Class Assignment 2: Estimating Pi using Monte Carlo

```python
import random
import matplotlib.pyplot as plt
```

```python
N = int(input("Enter the total no.of marbles to be placed: "))

circle_pts = 0          # Initializing the counts for points in circle and square
sqr_pts = 0
x = 0                   # variables for integration
y = 0
d = 0
L1 = []             
L2 = []                 # Storing the values of Pi

for i in range(1,N+1):
  L1.append(i)          # This is used for plotting purpose
  S = 0
  for j in range(50):   # For 50trials each
    circle_pts = 0
    sqr_pts = 0
    for k in range(i):              # Monte-Carlo Method
      x = random.uniform(-1,1)
      y = random.uniform(-1,1)
      d = x**2 + y**2
      if (d <= 1):              # Points under the integral valu are considered
        circle_pts += 1
      sqr_pts += 1    
    S += 4*circle_pts/sqr_pts
  S /= 50
  L2.append(S)

  
plt.plot(L1,L2)
plt.xlabel('N')
plt.ylabel('Pi value')
plt.title('Finding out value of Pi using Monte Carlo Simulation')
plt.show() 
```
