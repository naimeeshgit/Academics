# Class Assignment 8: Predator-Prey Model 

Simulate the pedator-prey model discussed in class today. <br>
$N_1(t)$ and $N_2(t)$ are the populations of the prey and predator respectively. <br>
The equations are: <br>
1. \begin{align*}
    \frac{dN_1(t)}{dt} = rN_1(t)\bigg(1 - \frac{N_1(t)}{K} \bigg) - \alpha N_1 N_2
\end{align*}
2. \begin{align*}
    \frac{dN_2(t)}{dt} = -c N_2(t) + \beta N_1 N_2
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def growth(y, t, r, c, k, alpha, beta):         # function for computing the equations of the predatr-prey model as shown above
    N1, N2 = y
    dydt = [r*N1*(1 - N1/k) - alpha*N1*N2, -c*N2 + beta*N1*N2]
    return dydt


r,alpha,c,beta,k = .5,1.2,.8,2.0,0.5            # initializing the constants
t = np.linspace(0, 29, 501)                     # time axis

y0 = [3,1]                                      # population of prey and predator
sol = odeint(growth, y0, t, args=(r,c,k,alpha,beta))        # solving differential equation with pre-defined function

plt.plot(t, sol[:, 0], 'b', label='prey')
plt.plot(t, sol[:, 1], 'g', label='predator')
plt.legend(loc='best')
plt.xlabel('t')
plt.title('Predator-Prey Model (Prey > Predator, in population)')
plt.grid()
plt.show()
```

```python
y0 = [3,3]          # Changing the populations for the three different cases    
sol = odeint(growth, y0, t, args=(r,c,k,alpha,beta))

plt.plot(t, sol[:, 0], 'b', label='prey')
plt.plot(t, sol[:, 1], 'g', label='predator')
plt.legend(loc='best')
plt.xlabel('t')
plt.title('Predator-Prey Model (Prey = Predator, in population)')
plt.grid()
plt.show()
```
